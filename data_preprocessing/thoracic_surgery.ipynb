```python
def true_false_converter(thoracic_data, column):
    thoracic_data.loc[thoracic_data[column] == 'T', column] = 1
    thoracic_data.loc[thoracic_data[column] == 'F', column] = 0
```

```python
true_false_converter(thoracic_data,"pain_before_b")
true_false_converter(thoracic_data,"haemoptysis_before_b")
true_false_converter(thoracic_data,"dyspnoea_before_b")
true_false_converter(thoracic_data,"cough_before_b")
true_false_converter(thoracic_data,"weakness_before_b")
true_false_converter(thoracic_data,"type_2_diabetes_b")
true_false_converter(thoracic_data,"mi_up_to_6_months_g")
true_false_converter(thoracic_data,"pad_b")
true_false_converter(thoracic_data,"smoking_b")
true_false_converter(thoracic_data,"asthma_b")
true_false_converter(thoracic_data,"survival_after_one_year_b")
```
