```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
ionosphere_data = pd.read_csv('../data/ionosphere.csv')
```

```python
ionosphere_data
```

```python
ionosphere_data = ionosphere_data.drop(columns=['a1_g'])
ionosphere_data = ionosphere_data.drop(columns=['a0_g'])
```

```python
# g is 1
ionosphere_data.loc[ionosphere_data['a34_g'] == 'g', 'a34_g'] = 1

# b is 0
ionosphere_data.loc[ionosphere_data['a34_g'] == 'b', 'a34_g'] = 0
```

```python
ionosphere_data
```

```python
ionosphere_export = np.array(ionosphere_data)
ionosphere_headers = np.array(ionosphere_data.columns)

%store ionosphere_export
%store ionosphere_headers
```

```python
x1 = ionosphere_export[:,1]
x2 = ionosphere_export[:,2]
x3 = ionosphere_export[:,3]
y = ionosphere_export[:,-1]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x1, y, color='r')
ax.scatter(x2, y, color='b')
ax.scatter(x3, y, color='g')
ax.set_title('First sensor data vs. class')
ax.set_xlabel('Sensor data')
ax.set_ylabel('Class')
plt.show()
```

```python
x1 = ionosphere_export[:,0]
x2 = ionosphere_export[:,1]
x3 = ionosphere_export[:,2]
x4 = ionosphere_export[:,3]
x5 = ionosphere_export[:,4]
x6 = ionosphere_export[:,5]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x1, x2, color='r')
ax.scatter(x3, x4, color='g')
ax.scatter(x5, x6, color='b')

ax.set_title('Correlation among sensor data')
ax.set_xlabel('Sensor point1')
ax.set_ylabel('Sensor point2')

```

```python
classes = ('Good', 'Bad')
y_pos = np.arange(len(classes))
num_good = np.count_nonzero(ionosphere_export[:,-1]==1)
num_bad = np.count_nonzero(ionosphere_export[:,-1]==0)
num = [num_good, num_bad]

plt.bar(y_pos, num, align='center', alpha=0.5)
plt.xticks(y_pos, classes)
plt.ylabel('Classes')
plt.title('Number of Good/Bad instances')

plt.show()
```
